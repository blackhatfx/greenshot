#  GBM Simulation in R

# clear all working space & Graphihcs

```r
rm(list = ls())
graphics.off()
```

# Define the GBM simulation function

```r
gbm.f <- function(n, S0, mu, sigma) {
  t <- 1
  t.s <- seq(0, t, length = n + 1)
  dt <- t / n
  Bt <- sqrt(dt) * cumsum(rnorm(n + 1, 0, 1))
  St <- S0 * exp((mu - sigma^2 / 2) * t.s + sigma * Bt)
  St
}
```

# Simulate GBM and store the results in 'rt'

```r
rt <- gbm.f(n = 1000, S0 = 10, mu = 0.05, sigma = 0.15)

# Plot the results as a line graph
plot(seq(0, 1, length = 1000 + 1), rt, type = "l", col = "blue",
     ylab = "Simulated Returns", xlab = "Time")
```
